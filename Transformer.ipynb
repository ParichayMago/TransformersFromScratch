```python
import numpy as np

def positional_encoding(seq_len, d_model, base=10000.0):
    """
    Return PE of shape (seq_len, d_model) using sin/cos formulation.
    Odd d_model -> last column is sin.
    """
    pe = np.zeros((seq_len, d_model), "float")

    position = np.arange(0, seq_len)[:, np.newaxis]# seq_len, 1
    div_terms = (base**(np.arange(0, d_model, 2)/d_model))[np.newaxis, :]

    terms = position/div_terms

    print(pe.shape, terms.shape)

    pe[: ,0::2] = np.sin(terms)
    pe[:, 1::2] = np.cos(terms)
    
    return pe

def add_positional_encoding(x, base=10000.0):
    """
    Add PE to input x of shape (B, T, d_model); return same shape.
    """
    B, T, d_model = x
    B = np.zeros((B, 1,1))

    pe = positional_encoding(T, d_model, base)
    print(pe.shape)
    
    return B + pe  
```
